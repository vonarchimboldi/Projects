```python
import pandas as pd
import numpy as np
import os
import sys
import glob
```

```python
os.chdir('../models')
```

```python
path = os.getcwd()
```

```python
files = glob.glob(os.path.abspath(path) + '/*.csv')
```

```python
files = [x for x in files if '_feat_imps_adj.csv' in x]
```

```python
li = []
for f in files:
    li.append(pd.read_csv(f))
    
df = pd.concat(li, axis = 1)
```

```python
df.columns = ['feature1', 'vif_score', 'feature2', 'hclust_score', 'feature3', 'mixed', 'feature', 'base']
```

```python
df.drop(['feature1', 'feature2', 'feature3'], axis = 1, inplace = True)
```

```python
cols = ['feature', 'base',  'hclust_score', 'vif_score', 'mixed']
df = np.round(df[cols].fillna(0), 2)
```

```python
df.to_csv('importance_comps_adj.csv', index = False)
```
